```python
#pandas to read data into python
#seaborn
import pandas as pd
```

```python
data=pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv',index_col=0)
data.head()
data.tail()
data.shape
```

# single columnin panda=series

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.pairplot(data,x_vars=['TV','Radio','Newspaper'],y_vars='Sales')
```

```python
sns.pairplot(data,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',size=7,aspect=0.7,kind='reg')
```

```python
#linear regression
```

```python
feature_cols=['TV','Radio','Newspaper']
X=data[feature_cols]
# orx=data[['TV','Radio','Newspaper']]
X.head()
print (type(X))
print(X.shape)
```

```python
y=data['Sales']
y.head()
print(type(y))
print(y.shape)
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, random_state=1)
# default =0.25 for test
```

```python
#Linerar regression
# import
from sklearn.linear_model import LinearRegression
#instatntiate
linreg=LinearRegression()
#fit
linreg.fit(X_train,y_train)
#predict
y_pred=linreg.predict(X_test)
```

```python
print(linreg.coef_)
print(linreg.intercept_)
```

```python
#pair feature names with coefficents
zip(feature_cols,linreg.coef_)
```

```python
# regression problems have a continuos response 
# we dont use classification accuracy
true=[100,90,50,20]
pred=[90,50,50,30]
```

```python
#MAE =mean of absolute value of errors
from sklearn import metrics
print (metrics.mean_absolute_error(true,pred))
```

```python
#MSE- 
print(metrics.mean_squared_error(true,pred))
#minimizes larger erros by punishing larger errors
```

```python
#RMSE
import numpy as np
print(np.sqrt(metrics.mean_squared_error(true,pred)))
#RMSE-more interpretable =same unit as y metric      
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
